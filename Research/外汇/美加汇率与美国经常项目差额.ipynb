```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'../data/美元兑加元.xls', index_col=0)
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(df[u'美元兑加元'], color='blue', label=u'美元兑加元')
ax1.legend(loc='upper left')
ax2 = ax1.twinx()
ax2.plot(df[u'美国:经常项目差额占GDP比重:季调'], color='red', label=u'美国:经常项目差额占GDP比重:季调')
ax2.legend()
```
