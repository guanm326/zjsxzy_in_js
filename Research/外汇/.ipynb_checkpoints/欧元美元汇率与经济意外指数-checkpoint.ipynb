```python
from WindPy import w
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
data = w.edb('G1137689,G0930000', '2003-01-01', '2018-05-22', 'Fill=Previous')
```

```python
econ_df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['us', 'eu'])
```

```python
econ_df['diff'] = econ_df['us'] - econ_df['eu']
```

```python
econ_df['diff'].plot(figsize=(16, 9))
```

```python
data = w.wsd('EURUSD.FX', 'close', '2003-01-01', '2018-05-22')
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['eurusd'])
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(econ_df['diff'])
ax2 = ax1.twinx()
ax2.plot(df['eurusd'], color='red')
# df['eurusd'].plot(figsize=(16, 9))
# econ_df['diff'].plot(figsize=(16, 9))
```
