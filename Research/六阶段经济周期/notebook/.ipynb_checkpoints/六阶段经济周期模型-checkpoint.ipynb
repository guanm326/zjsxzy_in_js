```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('../data/assets.xls', index_col=0)
```

```python
df.tail()
```

```python
rdf = df.rolling(120).mean()
```

```python
zdf = (rdf - rdf.mean()) / rdf.std()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(zdf['CCFI.WI'])
```
