```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('../data/assets.xls', index_col=0)
```

```python
df.tail()
```

```python
rdf = df.rolling(10).mean()
```

```python
zdf = (rdf - rdf.mean()) / rdf.std()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(zdf['NH0100.NHF'], label=u'商品')
ax.plot(zdf['881001.WI'], label=u'股票')
axx = ax.twinx()
ax.legend(loc='upper left', fontsize=15)
axx.plot(zdf[u'10Y国债'], label=u'债券(右轴)', color='red')
axx.legend(fontsize=15)
axx.invert_yaxis()
```

```python
zdf.to_excel(u'经济周期.xlsx')
```
