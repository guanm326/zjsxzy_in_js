```python
# Python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'./examples/六大电厂日均耗煤量.xlsx')
```

```python
df.head()
```

```python
df['ds'] = df.index
```

```python
df.columns = ['y', 'ds']
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(df['y'], label=u'六大电厂日均耗煤量')
ax.legend()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast.index = pd.to_datetime(forecast['ds'])
```

```python
fig = m.plot(forecast)
```

```python
fig = m.plot_components(forecast)
```
