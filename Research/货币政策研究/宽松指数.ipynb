# 宽松指数

从央行《货币政策执行报告》中提取宽松关键字，并计算词距离

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'./宽松.xlsx')
```

```python
df.head()
```

```python
loose = df[0].dropna()
```

```python
cpi = pd.read_excel('./CPI.xlsx', index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(loose.index, loose.rolling(4).mean(), color='blue')
ax1 = ax.twinx()
ax1.plot(cpi.index, cpi[u'CPI同比'], color='red')
# ax1.plot(cpi.index, cpi[u'PPI同比'], color='purple')
```

```python
cgb = pd.read_excel('./CGB.xlsx', index_col=0)
```

```python
cgb = cgb.resample('Q').last()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(loose.index, loose.rolling(4).mean(), color='blue', linewidth=5)
ax.set_title(u'央行宽松语气指数', size=20)
# ax1 = ax.twinx()
# ax1.plot(cgb.index, cgb['1y'], color='red')
# ax1.plot(cgb.index, cgb['10y'], color='purple')
```

```python
loose.rolling(4).mean().rank(pct=True).tail()
```
