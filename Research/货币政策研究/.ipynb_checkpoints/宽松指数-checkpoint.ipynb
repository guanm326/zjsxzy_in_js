# 宽松指数

从央行《货币政策执行报告》中提取宽松关键字，并计算词距离

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'./宽松.xlsx', sheetname='Sheet2')
```

```python
df.head()
```

```python
loose = df[u'宽松指数'].dropna()
gdp = df[u'GDP同比'].dropna()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(loose.index, loose, color='blue')
ax1 = ax.twinx()
ax1.plot(gdp.index, gdp, color='red')
```

```python
cpi = pd.read_excel('./CPI.xlsx', index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(loose.index, loose, color='blue')
ax1 = ax.twinx()
ax1.plot(cpi.index, cpi[u'CPI同比'], color='red')
ax1.plot(cpi.index, cpi[u'PPI同比'], color='purple')
```

```python
cgb = pd.read_excel('./CGB.xlsx', index_col=0)
```

```python
cgb = cgb.resample('Q').last()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(loose.index, loose, color='blue')
ax1 = ax.twinx()
ax1.plot(cgb.index, cgb['1y'], color='red')
# ax1.plot(cgb.index, cgb['10y'], color='purple')
```

```python
policy = pd.read_excel(u'./政策-收紧.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.bar(policy.index, policy[policy.columns[0]], color='red', width=10)
ax1 = ax.twinx()
ax1.plot(cgb.index, cgb['1y'], color='blue')
# ax1.plot(cgb.index, cgb['10y'], color='purple')
```
