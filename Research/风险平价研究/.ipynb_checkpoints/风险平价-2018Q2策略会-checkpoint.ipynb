```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import empyrical
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
def metric(daily_ret, period='daily'):
    an_ret = empyrical.annual_return(daily_ret, period=period)
    sharpe = empyrical.sharpe_ratio(daily_ret, period=period)
    vol = empyrical.annual_volatility(daily_ret, period=period)
    maxdraw = empyrical.max_drawdown(daily_ret)
    print 'Annual return: %.2f%%'%(an_ret*100)
    print 'Sharpe ratio: %.2f'%(sharpe)
    print 'Annual volatility: %.2f%%'%(vol*100)
    print 'Max drawdown: %.2f%%'%(maxdraw*100)
```

# 读入数据并处理

```python
fname = '../data/assets.xlsx'
```

```python
odf = pd.read_excel(fname, index_col=0)
```

```python
odf = odf.replace([0.0], np.NAN)
```

```python
odf.head()
```

```python
df = odf[['881001.WI', 'CBA00101.CS', 'CCFI.WI']]
```

```python
df.dropna().head()
```

# 计算收益率和波动率

```python
ret_df = df.pct_change().dropna()
vol_df = ret_df.rolling(window=243).apply(lambda x: empyrical.annual_volatility(x, period='daily'))
```

```python
metric(ret_df['881001.WI'])
```

# 计算风险平价组合

```python
vol_df = vol_df.dropna()
```

```python
months = pd.Series([(x.year, x.month) for x in vol_df.index], index=vol_df.index)
rebalance_dates = months.drop_duplicates(keep='last').index
```

```python
rebalance_dates
```

```python
weight = pd.DataFrame(columns=vol_df.columns, index=vol_df.index)
wb = np.zeros(vol_df.shape[1])
for date in vol_df.index:
    weight.loc[date] = wb
    if date in rebalance_dates:
        vol = vol_df.loc[date]
        wb = 1. / vol
        wb /= wb.sum()
```

```python
(1 + (weight * ret_df)).cumprod().dropna().plot(figsize=(16, 9))
```

```python
portfolio_ret1 = (weight * ret_df).sum(axis=1)
portfolio_ret1 = portfolio_ret1[portfolio_ret1 != 0]
```

```python
metric(portfolio_ret1)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret1).cumprod(), label=u'风险平价组合', color='blue')
plt.legend()
```

# 债券加杠杆

```python
leverage = 2.
weight2 = weight.copy()
weight2['CBA00101.CS'] *= leverage
```

```python
portfolio_ret2 = (weight2 * ret_df).sum(axis=1)
portfolio_ret2 = portfolio_ret2[portfolio_ret2 != 0]
```

```python
metric(portfolio_ret2)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret2).cumprod(), label=u'100%杠杆组合', color='blue')
plt.plot((1+portfolio_ret1).cumprod(), label=u'无杠杆组合', color='red')
plt.legend()
```

# 与股票的对比

```python
plt.figure(figsize=(16, 9))
plt.plot((1+ret_df[ret_df.index >= '2003-01-29']['881001.WI']).cumprod(), label=u'万得全A', color='red')
plt.plot((1+portfolio_ret2).cumprod(), label=u'100%杠杆组合', color='blue')
plt.legend()
```

# 增加股票配置比例

```python
weight3 = weight.copy()
weight3['CBA00101.CS'] -= weight3['881001.WI']
weight3['881001.WI'] *= 3
```

```python
portfolio_ret3 = (weight3 * ret_df).sum(axis=1)
portfolio_ret3 = portfolio_ret3[portfolio_ret3 != 0]
```

```python
metric(portfolio_ret3)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret3).cumprod(), label=u'增加股票持仓组合', color='blue')
plt.plot((1+portfolio_ret1).cumprod(), label=u'无杠杆组合', color='red')
plt.legend()
```

# 规避股市流动性风险

```python
data = w.wsd('881001.WI', 'volume,close', '2002-01-01', '2018-05-31')
```

```python
pdf = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=['volume', 'close'])
```

```python
pdf['ret'] = pdf['close'].pct_change()
```

```python
pdf['liquidity'] = pdf['ret'].abs() * 1e8 / pdf['volume']
```

```python
pdf['z'] = pdf['liquidity'].rolling(window=243).apply(lambda x: (x[-1] - x.mean()) / x.std())
```

```python
pdf.loc[pdf['z'].abs() < 2, 'z'] = 0
```

```python
plt.figure(figsize=(16, 9))
plt.bar(pdf.index, pdf['z'], width=8)
```

```python
money_stock = (pdf['z'].fillna(0) != 0).rolling(window=10).sum()
```

```python
money_stock = money_stock.fillna(0)
```

```python
index = money_stock[money_stock != 0].index
```

```python
weight4 = weight3.copy()
```

```python
skip = weight4.loc[index.intersection(weight4.index), '881001.WI']
weight4.loc[index.intersection(weight4.index), '881001.WI'] = 0.
money_stock.loc[skip.index] = skip
```

```python
(1 + weight4['881001.WI'] * ret_df['881001.WI']).cumprod().plot(figsize=(16, 9))
```

```python
money_ret = odf['H11025.CSI'].pct_change()
extra_ret_stock = money_ret * money_stock
extra_ret_stock = extra_ret_stock.fillna(0)
```

```python
portfolio_ret4 = (weight4 * ret_df).sum(axis=1)
portfolio_ret4 = portfolio_ret4[portfolio_ret4 != 0]
portfolio_ret4 += extra_ret_stock.loc[portfolio_ret4.index]
```

```python
metric(portfolio_ret4)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret4).cumprod(), label=u'规避股市流动性风险组合', color='green')
plt.plot((1+portfolio_ret3).cumprod(), label=u'增加股票持仓组合', color='blue')
plt.plot((1+portfolio_ret1).cumprod(), label=u'无杠杆组合', color='red')
plt.legend()
```
