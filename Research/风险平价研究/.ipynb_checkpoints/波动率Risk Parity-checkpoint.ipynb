```python
def metric(daily_ret, period='daily'):
    an_ret = empyrical.annual_return(daily_ret, period=period)
    sharpe = empyrical.sharpe_ratio(daily_ret, period=period)
    vol = empyrical.annual_volatility(daily_ret, period=period)
    maxdraw = empyrical.max_drawdown(daily_ret)
    print 'Annual return: %.2f%%'%(an_ret*100)
    print 'Sharpe ratio: %.2f'%(sharpe)
    print 'Annual volatility: %.2f%%'%(vol*100)
    print 'Max drawdown: %.2f%%'%(maxdraw*100)
```

```python
# df = odf[['SPX Index', # 股
#           'LQD.P', # 债
#           'SPGSCITR.SPI', # 商品总指数
# #           'SPTAUUSDOZ.IDC', 'B.IPE', # 商品
#           'USDX.FX', # 外汇
#           ]]
df = odf[['000300.SH', 'CBA00101.CS', 'AU9999.SGE', 'CCFI.WI']]
# df = odf[['SP500.SPI', 'LBUSTRUU Index', 'SPGSCITR.SPI', 'BBREIT Index']]
```

```python
# ret_df = df.pct_change().dropna()
# vol_df = ret_df.rolling(window=243).apply(lambda x: empyrical.annual_volatility(x, period='daily'))
ret_df = df.resample('M').last().pct_change().dropna()
vol_df = ret_df.rolling(window=12).apply(lambda x: empyrical.annual_volatility(x, period='monthly'))
```

# 加入房价

```python
house = pd.read_excel(u'../data/70个大中城市二手住宅价格指数(月).xls')
```

```python
ret_df = ret_df.merge(house, how='outer', left_index=True, right_index=True).dropna()
```

```python
ret_df = ret_df[ret_df.index >= '2009-12-01']
```

```python
house_vol = house.rolling(window=12).apply(lambda x: empyrical.annual_volatility(x, period='monthly'))
```

```python
vol_df = vol_df.merge(house_vol, how='outer', left_index=True, right_index=True).dropna()
```

# ======================================================

```python
vol_df = vol_df[vol_df.index >= '2009-12-01']
```

```python
months = pd.Series([(x.year, x.month) for x in vol_df.index], index=vol_df.index)
rebalance_dates = months.drop_duplicates(keep='last').index
```

```python
rebalance_dates
```

```python
weight = pd.DataFrame(columns=vol_df.columns, index=vol_df.index)
wb = np.zeros(vol_df.shape[1])
for date in vol_df.index:
#     print date, wb
    weight.loc[date] = wb
    if date in rebalance_dates:
        vol = vol_df.loc[date]
        wb = 1. / vol
        wb /= wb.sum()
#         print date, vol, wb
```

```python
weight.drop_duplicates().to_excel(u'C:/Users/zhangyang01/Desktop/RP(中国包括房价).xlsx')
```

```python
(1 + (weight * ret_df)).cumprod().dropna().plot(figsize=(16, 9))
```

```python
portfolio_ret = (weight * ret_df).sum(axis=1)
```

```python
portfolio_ret = portfolio_ret[portfolio_ret != 0]
```

```python
portfolio_ret.tail()
```

```python
(1+portfolio_ret).cumprod().to_csv('C:/Users/zhangyang01/Desktop/nav.csv')
```

```python
metric(portfolio_ret, period='monthly')
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret).cumprod(), label='volatility risk parity')
plt.legend()
```

# 未来波动率

```python
weight = pd.DataFrame(columns=ret_df.columns, index=ret_df.index)
wb = np.zeros(ret_df.shape[1])
for date in ret_df.index:
    weight.loc[date] = wb
    if date in rebalance_dates:
        past_return = ret_df[ret_df.index >= date][121:]
        vol = past_return.std()
        wb = 1. / vol
        wb /= wb.sum()
```

```python
portfolio_ret2 = (weight * ret_df).sum(axis=1)
```

```python
portfolio_ret2 = portfolio_ret2[portfolio_ret2 != 0]
```

```python
metric(portfolio_ret2)
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1+portfolio_ret).cumprod(), label='volatility risk parity')
plt.plot((1+portfolio_ret2).cumprod(), label='real volatility risk parity')
plt.legend()
```

```python
df = pd.DataFrame({'past': (1+portfolio_ret).cumprod(), 'future': (1+portfolio_ret2).cumprod()})
```
