```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.semi_supervised import label_propagation
%matplotlib inline
```

```python
df = pd.read_excel('./data/stock price.xlsx')
```

```python
df.head()
```

```python
X = pd.DataFrame(index=df.index, columns=df.columns)
for date in df.index:
    past = df[df.index <= date][-500:]
    X.loc[date] = (df.loc[date] - past.mean()) / past.std()
```

```python
plt.figure(figsize=(16, 9))
X['881001.WI'].plot()
```

```python
X_month = X.rolling(window=20).mean().resample('M').last()
```

```python
X_month = X_month.dropna()
```

```python
plt.figure(figsize=(16, 9))
X_month['881001.WI'].plot()
```

```python
# æƒ…æ™¯ï¼šå‘¨æœŸè‚¡ç‰›å¸‚
scenario = X_month[((X_month.index >= '2009-01-01') & (X_month.index <= '2009-07-01')) |
                   ((X_month.index >= '2010-07-01') & (X_month.index <= '2010-10-01'))]
```

```python
X_month['label'] = -1
```

```python
X_month.loc[X_month.index <= '2012-01-01', 'label'] = 0
X_month.loc[scenario.index, 'label'] = 1
```

```python
X_month.tail()
```

```python
X_, y = X_month[X.columns.difference(['label'])], X_month['label']
```

```python
X_
```

```python
clf = label_propagation.LabelSpreading().fit(X_, y)
```

```python
clf.predict(X_)
```

```python
pred = clf.predict_proba(X_)[:, 1]
```

```python
plt.figure(figsize=(16, 9))
plt.plot(pd.Series(pred, index=X_.index))
```
