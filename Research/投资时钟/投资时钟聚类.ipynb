# GDP、CPI投资时钟（美林时钟）

```python
df = pd.DataFrame({'cpi': macro_df['CPI'][gdp_df.index], 'gdp': gdp_df['GDP']}, index=gdp_df.index)
```

```python
df.plot()
```

```python
change_df = df.pct_change(periods=3)
```

```python
change_df.plot()
```

```python
sns.lmplot('cpi', 'gdp', data=change_df, fit_reg=False)
```
