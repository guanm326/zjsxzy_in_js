```python
raw_data = w.wsd(['CI005021.WI', '881001.WI'], 'close', '2006-12-31', '2016-12-31')
```

```python
sdf = pd.DataFrame(np.array(raw_data.Data).T, index=raw_data.Times, columns=raw_data.Codes)
```

```python
excess_return = sdf['CI005021.WI'].pct_change() - sdf['881001.WI'].pct_change()
```

```python
plt.figure(figsize=(16, 9))
plt.plot((1 + excess_return).cumprod() - 0.5)
plt.plot(r_ch[r_ch.index >= '2006-01-01'])
```
