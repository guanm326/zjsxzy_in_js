```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
DATA_DIR = 'D:/Data/index/'
```

```python
files = [f.rstrip('.xlsx') for f in os.listdir(DATA_DIR) if f.startswith('CI')]
```

```python
dic = {}
for f in files:
    fname = '%s/%s.xlsx'%(DATA_DIR, f)
    temp = pd.read_excel(fname, index_col=0)
    dic[f] = temp['volume']
df = pd.DataFrame(dic)
```

```python
df.head()
```

```python
df = df.div(df.sum(axis=1), axis='index')
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df['CI005021.WI'].rolling(window=20).mean())
```
