```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(u'C:/Users/zhangyang01/Dropbox/嘉实/Excel数据分析/PB-ROE.xlsx', sheetname='Sheet2')
```

```python
df.head()
```

```python
df = df[(df['PB'] > 0) & (df['ROE'] > 0)]
df = df[df['PB'] <= 15]
df = df[df['ROE'] <60]
```

```python
df.head()
```

```python
z = np.polyfit(df['ROE'], df['PB'], deg=1)
```

```python
z
```

```python
p = np.poly1d(z)
```

```python
y = p(df['ROE'])
```

```python
plt.figure(figsize=(16, 9))
plt.scatter(df['ROE'], df['PB'])
plt.plot(df['ROE'], y, '-k')
```
