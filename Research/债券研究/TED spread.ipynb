```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
def wind2df(raw_data):
    dic = {}
    for data, code in zip(raw_data.Data, raw_data.Codes):
        dic[str(code)] = data
    return pd.DataFrame(dic, index=raw_data.Times)
```

```python
data = w.wsd("SHIBOR3M.IR,CGB3M.WI", "ytm_b", "2017-06-07", "2017-07-06", "returnType=1")
```

```python
df = wind2df(data)
```

```python
df.plot()
```
