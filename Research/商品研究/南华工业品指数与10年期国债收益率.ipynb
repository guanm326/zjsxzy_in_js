```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'./南华工业品指数.xlsx', index_col=0)
```

```python
df = df.fillna(method='ffill')
```

```python
df.head()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax2 = ax1.twinx()
ax1.plot(df.index, df[u'南华工业品指数'], color='blue', label=u'南华工业品指数')
ax2.plot(df.index, df[u'10年期国债收益率'], color='red', label=u'10年期国债收益率')
ax1.legend()
ax2.legend(loc='upper left')
```
