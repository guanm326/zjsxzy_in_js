```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'D:/Data/assets/commodity/yiwu/总指数.xlsx', index_col=0)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df[u'历史分位'] = df[u'价格指数'].rank(pct=True)
df[u'历史分位5日均值'] = df[u'历史分位'].rolling(window=5).mean()
```

```python
df[u'历史分位5日均值'].plot(figsize=(16, 9))
```
