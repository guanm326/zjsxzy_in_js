```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import empyrical
from WindPy import w
%matplotlib inline
```

```python
w.start()
```

```python
def wind2df(raw_data):
    return pd.DataFrame({'close': raw_data.Data[0]}, index=raw_data.Times)
```

```python
data = w.wsd('881001.WI', 'close', '2000-01-01', '2017-11-30')
```

```python
base = wind2df(data)
base.index = pd.to_datetime(base.index)
```

```python
df = pd.read_excel('./data/theme.xlsx')
```

# 一定时期各行业收益率

```python
start_date, end_date = '2011-11-01', '2012-05-01'
```

```python
sample_df = df[(df.index >= start_date) & (df.index <= end_date)]
sample_base = base[(base.index >= start_date) & (base.index <= end_date)]
```

```python
ret = (sample_df.iloc[-1] - sample_df.iloc[0]) / sample_df.iloc[0]
base_ret = (sample_base.iloc[-1]['close'] - sample_base.iloc[0]['close']) / sample_base.iloc[0]['close']
```

```python
ret_df = pd.DataFrame({u'绝对收益率': ret, u'超额收益率': ret - base_ret})
```

```python
ret_df.sort_values(u'超额收益率')[::-1].to_excel('./data/3.xlsx')
```
