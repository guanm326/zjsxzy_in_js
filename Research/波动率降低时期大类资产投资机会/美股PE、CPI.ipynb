```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('./data/pe.xlsx', index_col=0)
```

```python
df.head()
```

```python
df.sort_values('PE')
```

```python
plt.figure(figsize=(16, 9))
df['PE'].plot()
```

```python
plt.figure(figsize=(16, 9))
s, e = '1908', '1921'
df[df.index <= '%s-01-01'%(s)]['PE'].plot(c='b', label='')
df[df.index >= '%s-01-01'%(e)]['PE'].plot(c='b', label='')
df[(df.index >= '%s-01-01'%(s)) & (df.index <= '%s-01-01'%(e))]['PE'].plot(c='r', label='%s-%s'%(s, e))
plt.title('PE')
plt.legend()
```
